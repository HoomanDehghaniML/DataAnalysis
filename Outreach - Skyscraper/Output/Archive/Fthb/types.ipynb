```python
import pandas as pd
import numpy as np
```

```python
# load dataframes
df = pd.read_xml("https://www.squareone.ca/page-sitemap.xml")
df_temp = pd.read_xml("https://www.squareone.ca/page-sitemap2.xml")

```

```python
# concat them
df = df.drop(columns=['image', 'changefreq', 'priority'])
df = pd.concat([df, df_temp])
df = df.reset_index()
df = df.drop(columns=['lastmod', 'image', 'index'])
df = df.rename(columns={"loc": "URL"})
```

```python
# Add the type columns
df.loc[:, ['Priority', 'Traffic','Type1', 'Type2', 'Type3', 'Type4', 'Type5', 'Type6']] = ""
```

```python
# Populate Type1: auto or home
df.loc[:, 'Type1'] = 'home'
```
